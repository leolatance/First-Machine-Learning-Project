```python
import pandas as pd

df = pd.read_csv('manutencao.csv')
df
```

```python
df.plot(kind="scatter", x="temperatura_forno", y="resistencia_aco")
```

```python
from sklearn import linear_model

modelo = linear_model.LinearRegression()
x = df[["temperatura_forno"]]
y = df[["resistencia_aco"]]

modelo.fit(x, y)

```

```python
print(modelo.predict([[3000]])[0][0])
```
